```python
import requests
import pandas as pd
```

```python
# closeされたPRのみ取得

# while文でパラメータを取得する
df = pd.DataFrame()
i = 1

while True:
  
  # ループしたい数式
  ## dfを取得する
  URL = 'https://api.github.com/repos/clustervr/ClusterONE/pulls?'
  headers = {'Authorization': 'token '+ token} # 取得したアクセストークン
  params = {
      'state':'closed',
      'sort':'updated',
      'per_page':100,
      'page': i ,
  }
  r = requests.get(URL, headers=headers, params=params)
  x = pd.json_normalize(r.json())

  df = pd.concat([df, x], ignore_index=True)
  
  i = i + 1
  print(i)

  if len(x.index) < 100:
    break

  ## dfを結合する

print(len(df.index))
```
